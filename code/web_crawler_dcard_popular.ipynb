```python
import pandas as pd
import requests
from requests_html import HTML
import re
from fake_useragent import UserAgent
import random
import time
ua = UserAgent()
```

```python
def Crawl(ID):
    link = 'https://www.dcard.tw/service/api/v2/posts/' + str(ID)
    requ = requests.get(link, headers = {
        "user-agent": ua.random
      })

    print(requ)
    # response 429: too many requests
    if requ.status_code==429:
        time.sleep(10)
#         Crawl(ID)
    if requ.status_code!=200:
#         time.sleep(0.01)
#         Crawl(ID)
        return pd.DataFrame(
            data=[0],
            columns=['id'])
    
    rejs = requ.json()

    return(pd.DataFrame(
        data=
        [{'id':rejs['id'],
          'title':rejs['title'],
          'excerpt':rejs['excerpt'],       #æ‘˜è¦
          'content':rejs['content'],
          'createdAt':rejs['createdAt'],              
          'anonymousSchool':rejs['anonymousSchool'],     #åŒ¿åæ ¡å
          'anonymousDepartment':rejs['anonymousDepartment'],    #åŒ¿åç³»å
          'gender':rejs['gender'],
          'forumName':rejs['forumName'],     #çœ‹æ¿åç¨±(ä¸­æ–‡)
          'forumAlias':rejs['forumAlias'],   #çœ‹æ¿åç¨±(è‹±æ–‡)
          'topics':rejs['topics'],           #æ¨™è¨˜
          'commentCount':rejs['commentCount'],
          'likeCount':rejs['likeCount'],}],
        columns=['id','title','excerpt','content','createdAt', 'anonymousSchool','anonymousDepartment',
                 'gender','forumName','forumAlias','topics','commentCount','likeCount']))
```

```python
# url = 'https://www.dcard.tw/_api/forums/makeup/posts?popular=true&limit=20'
url = 'https://www.dcard.tw/_api/posts?popular=true&limit=60'
resq = requests.get(url)
rejs = resq.json()

df = pd.DataFrame()
# len(rejs)
for i in range(30,len(rejs)):
    new_df = Crawl(rejs[i]['id'])
    if(new_df.at[0,'id']!=0):
        df = df.append(new_df)
#     time.sleep(1)
print(df.shape)
df
```
